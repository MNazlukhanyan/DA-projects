# Функции для работы

```python
# Обновленная функция для построения сводных таблиц и тепловых карт, анализа и сравнения характеристик
def generate_pivot_heatmap(df, index, columns, aggfunc='median'):
    """
    Функция создает сводные таблицы и тепловые карты, а также вычисляет и 
    отображает нормализованные медианные значения и количество заполненных значений по каждой комбинации параметров.
    
    Аргументы:
    - df -  датафрейм для анализа.
    - index - столбец, который будет использоваться в качестве индекса сводной таблицы.
    - columns - столбец, который будет использоваться в качестве столбцов сводной таблицы.
    - aggfunc - функция для расчета значений в сводной таблице, по умеолчанию медиана.
    
    Вывод:
    - pivot_table_med_percent - нормализованная сводная таблица с медианными значениями.
    - pivot_table_count - сводная таблица, показывающая количество заполненных значений.
    """
    
    # 1.1 построение сводной таблицы по медиане 'balance' для анализа
    pivot_table_median = df[df['balance'].notna()].pivot_table(
        values='balance',
        index=index,
        columns=columns,
        aggfunc=aggfunc
    )
    # 1.2 нормализация значений сводной таблицы с медианным значением
    pivot_table_med_percent = pivot_table_median.div(pivot_table_median.sum(axis=0), axis=1)

    
    # 2 создание сводной таблицы с количеством заполненных значений
    pivot_table_count_notna = df[df['balance'].notna()].pivot_table(
        values='balance',
        index=index,
        columns=columns,
        aggfunc='count'
    )

    # 3 создание сводной таблицы с долями  пропущенных значений
    missing_count_table = df.pivot_table(
        values='balance',
        index=index,
        columns=columns,
        aggfunc=lambda x: x.isnull().sum()
    )

    # 3.1: вычисление общего количества пропусков
    total_missing_count = df['balance'].isnull().sum()
    
    # 3.2: деление на общее количество пропусков
    missing_ratio_table = missing_count_table / total_missing_count
    
    # 3.3: нормализация по столбцам
    missing_ratio_table = missing_ratio_table.div(missing_ratio_table.sum(axis=0), axis=1)
    
    
    # присвоение динамического имени переменной в зависимости от аргументов
    pivot_table_name = f"{index}_{columns}"

    # 4. вывод сводных таблиц
    print(f"Медианные значения 'balance' по параметрам {index} и {columns}:")
    display(pivot_table_med_percent)

    print(f"Количество заполненных значений 'balance' по параметрам {index} и {columns}:")
    display(pivot_table_count_notna)

    print(f"Доля пропущенных значений 'balance' по параметрам {index} и {columns}:")
    display(missing_ratio_table)

    # 5. построение тепловых карт по полученным сводным таблицам
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))

    sns.heatmap(pivot_table_med_percent, annot=True, cmap='OrRd', fmt=".2f", cbar=True, ax=axes[0])
    axes[0].set_title(f'Процентное соотношение медианных значений {index} и {columns}')
    axes[0].set_xlabel(columns)
    axes[0].set_ylabel(index)

    sns.heatmap(pivot_table_count_notna, annot=True, cmap='OrRd', fmt=".0f", cbar=True, ax=axes[1])
    axes[1].set_title(f'Количество заполненных значений по параметрам {index} и {columns}')
    axes[1].set_xlabel(columns)
    axes[1].set_ylabel(index)

    sns.heatmap(missing_ratio_table, annot=True, cmap='OrRd', fmt=".2f", cbar=True, ax=axes[2])
    axes[2].set_title(f'Доля пропущенных значений по параметрам {index} и {columns}')
    axes[2].set_xlabel(columns)
    axes[2].set_ylabel(index)

    plt.tight_layout()
    plt.show()
    
    return pivot_table_med_percent, pivot_table_count_notna, missing_ratio_table
```

```python
# верхняя функция заполнения пропусков
def fill_balance_na(df, col1, col2):
    
    # создаем группы по важным переменным
    groups = df.groupby([col1, col2])
    
    # нижняя функция для заполнения пропусков медианой по группам
    def fill_na_with_group_median(group):
        median_balance = group['balance'].median()  # Находим медиану по группе
        group['balance'] = group['balance'].fillna(median_balance)  # Заполняем пропуски
        return group
    
    # заполняем по группам значения
    df_filled = groups.apply(fill_na_with_group_median).reset_index(drop=True)
    
    return df_filled
```

```python
def average_churn_by_category(df, categorical_columns, categorical_bool):
    # создадим словарь, в котором будут храниться средние уровни отточности по каждому из признаков
    churn_means = {}  
    # цикл проходит по дискретным признакам
    for col in categorical_columns:
        # условие, в котором проверяются дискретные булевые признаки (0, 1)
        if col in categorical_bool:
            # для таких столбцов по значению True
            filtered_bool = df[df[col] == 1]
            # считаем средний уровень отточности
            churn_mean = filtered_bool['churn'].mean()
        else:
            # считаем средний уровень отточности
            churn_mean = df.groupby(col)['churn'].mean().mean()
        
        # округление
        mm = round(churn_mean, 3)
        
        # каждое значение по признаку будет храниться в виде переменной mm_"название столбца"
        churn_mean_name = f"mm_{col}"
        churn_means[churn_mean_name] = mm

        # выводим средний уровень оттока для текущего признака
        print(f"Средний уровень оттока по признаку {col}:\n{mm}")
    
    return churn_means
```

```python
# Функция для отрисовки гистограммы по каждому непрерывному признаку
def hist_box_category_churn(df, column, title, legend_title, ylabel, xlabel, ticks=None, bins=None):
    sns.set(style="whitegrid")
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    if bins is None:
        bins = round(mt.log(df[column].max()) + 1)
    
    # Отрисовка гистограммы
    sns.histplot(
        data=df, 
        x=column, 
        stat='density', 
        hue='churn', 
        palette = ['#ff5252', '#90EE90'], 
        common_norm=False, 
        bins=bins,
        ax=axes[0],
        kde=True
    )

    # Оформление гистограммы
    axes[0].set_title(f'Распределение {title} по отточности')
    axes[0].set_ylabel(ylabel)
    axes[0].legend(['Ушёл', 'Не ушёл'], title = legend_title, loc='upper right', fontsize='medium')
    axes[0].set_xlabel(xlabel)
    if ticks:
        axes[0].xticks(range(df[column].min(), df[column].max() + 1, ticks))

    # Отрисовка боксплота
    sns.boxplot(
        x=column, 
        y='churn', 
        data=df,
        ax=axes[1],
        orient='h',
        palette=['#ff5252', '#90EE90']
    )

    # Оформление графика
    axes[1].set_title(f'Распределение {title} по отточности', fontsize=15)
    axes[1].set_xlabel(xlabel, fontsize=13)
    axes[1].set_ylabel(ylabel, fontsize=13)

    # Построение визуализаций
    plt.tight_layout
    plt.show()
```

```python
# Функция для отрисовки barplot и создания сводной таблицы для обычных категориальных признаков
def plot_pivot_category_churn(df, column, title, ylabel, xlabel, churn_means=None):
    # 1. Создание сводной таблицы с помощью pivot_table
    pivot = pd.pivot_table(
        df, 
        values='churn', 
        index=column, 
        aggfunc=['sum', 'count'], 
        margins=False
    )

    # 2. Переименование столбцов для удобства восприятия
    pivot.columns = ['Ушли', 'Количество значений']
    
    # 3. Подсчёт дополнительных метрик
    pivot['Не ушли'] = pivot['Количество значений'] - pivot['Ушли']
    pivot['Средний уровень оттока'] = pivot['Ушли'] / pivot['Количество значений']

    # 4. Перестановка столбцов в нужном порядке
    pivot = pivot[['Не ушли', 'Ушли', 'Количество значений', 'Средний уровень оттока']]

    # Создадим фильтрацию по минимальному порогу 1% от максимума количества значений 
    max_count = pivot['Количество значений'].sum()
    threshold = max_count * 0.01  

    pivot = pivot[pivot['Количество значений'] > threshold]
    
    # сортировка значений по среднему уровню оттока
    pivot = pivot.sort_values(by='Средний уровень оттока', ascending=False)

    # порядок индексов в зависимости от сортировки
    pivot_order = pivot.index.tolist()
    
    # 5. Отображение сводной таблицы
    print("\nСводная таблица:")
    display(pivot)

    # 6. Построение barplot
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))

    # Получаем среднее значение оттока для текущей категории из словаря churn_means, если задано
    category_mean = churn_means.get(f"mm_{column}", None) if churn_means else None

    # Строим barplot
    sns.barplot(
        x=pivot.index,
        y=pivot['Средний уровень оттока'],
        palette='Set2',
        edgecolor='black',
        order=pivot_order,
    )
    
    #sns.scatterplot(
    #    data=pivot.reset_index(),
    #    x='Средний уровень оттока',  
    #    y=df,        
    #    hue=column,
    #    palette='Set2',
    #    s=150
    #)
    
    # Добавляем текст на столбцы с фактическими значениями среднего оттока
    for index, value in enumerate(pivot['Средний уровень оттока']):
        plt.text(
            index,
            value - (pivot['Средний уровень оттока'].max() * 0.06),
            f'{value:.3f}',
            ha='center',
            va='bottom',
            fontsize=15
        )

    # Добавление линии среднего значения оттока по всей категории
    if category_mean is not None:
        plt.axhline(
            category_mean,
            color='red',
            linestyle='--',
            linewidth=2,
            label=(f'Средний отток по категории: {category_mean:.2f}')
        )

    # Оформление графика
    plt.title(f'Средний уровень оттока по категории {title}', fontsize=15)
    plt.ylabel(ylabel, fontsize=13)
    plt.xlabel(xlabel, fontsize=13)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plot_boolean_churn(df, column, title):
    # Расчёт среднего уровня оттока для каждой категории булевого признака
    mean_churn = df.groupby(column)['churn'].mean()
    
    # Построение barplot
    sns.set(style="whitegrid")
    plt.figure(figsize=(8, 5))
    sns.barplot(
        x=mean_churn.index, 
        y=mean_churn.values, 
        palette=['#ff5252', '#90EE90'],
        edgecolor='black'
    )

    for index, value in enumerate(mean_churn):
        plt.text(
            index, 
            value - 0.02, 
            f'{value:.3f}', 
            ha='center', 
            va='bottom',
            fontsize=15
        )
    
    # Оформление графика
    plt.title(f'Средний уровень оттока в зависимости от {title}')
    plt.xlabel(f'Значения {title}', fontsize=14)
    plt.ylabel('Средний уровень оттока', fontsize=13)
    plt.xticks([0, 1], ['Нет', 'Есть'], fontsize=13)
    plt.show()
```

```python
# Функция для проведения теста Манна-Уитни
def mann_whitney_test(group1, group2, alpha=0.05, alternative='two-sided'):
    stat, p_value = stats.mannwhitneyu(group1, group2, alternative=alternative)
    if p_value < alpha:
        print(f"Манна-Уитни: Отвергаем нулевую гипотезу: между долями есть значимая разница. (p-value = {p_value})")
    else:
        print(f"Манна-Уитни: Нулевую гипотезу отвергнуть не удалось, разница незначима. (p-value = {p_value})")
    return p_value
```

```python
# Функция для проверки нормальности
def normality_test(data, alpha=0.05):
    stat, p_value = stats.shapiro(data)
    if p_value > alpha:
        print(f"Нормальность подтверждена (p-value = {p_value})")
        return True
    else:
        print(f"Нормальность отклонена (p-value = {p_value})")
        return False
```

```python
# Функция для проведения параметрического t-теста
def t_test(group1, group2, alpha=0.05):
    stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Используем параметр equal_var=False для неравных дисперсий
    if p_value < alpha:
        print(f"T-тест: Отвергаем нулевую гипотезу: между долями есть значимая разница. (p-value = {p_value})")
    else:
        print(f"T-тест: Нулевую гипотезу отвергнуть не удалось, разница незначима. (p-value = {p_value})")
    return p_value
```

# Проверка статистических гипотез

Исходя из полученных данных по признакам связанных с отточностью выдвинем несколько гипотез:

## Гипотеза о количестве продуктов и оттоке

**Гипотеза**: Односторонняя  (A > B)

- **Нулевая гипотеза (H₀)**: Среднее количество продуктов, которыми пользуются ушедшие и оставшиеся клиенты, равно.
- **Альтернативная гипотеза (H₁)**: Среднее количество продуктов у ушедших клиентов больше, чем у оставшихся.

Установим критический уровень статистической значимости 5%

```python
# Отфильтруем значения по двум группам
products_churn_true = df_filtered.query('churn == True')['products']
products_churn_false = df_filtered.query('churn == False')['products']

# Проведем проверку на нормальность распределения
p_value_products_normality = normality_test(df_filtered['products'])
if p_value_products_normality > 0.05:
    p_value_products = t_test(products_churn_true, products_churn_false)
    print(f"p-value : {p_value_products}")
else:
    print("Распределение количества продуктов не является нормальным, лучше использовать непараметрический тест.")
```

1. **Тип теста**: Непараметрический
2. **Тест**: Тест Манна-Уитни

```python
# Выполним тест Манна-Уитни
p_value_products = mann_whitney_test(products_churn_true, products_churn_false, alternative='greater')

# Посчитаем отношение среднего кол-ва продуктов ушедших к оставшимся
products_difference = products_churn_true.mean() / products_churn_false.mean()

print(f'Отношение среднего кол-ва продуктов ушедших к оставшимся {products_difference}')
if products_difference > 1:
    print('Среднее кол-во продуктов ушедших клиентов больше, чем у оставшихся клиентов.')
if products_difference < 1:
    print('Среднее кол-во продуктов ушедших клиентов меньше, чем у оставшихся')
if products_difference == 1:
    print('Среднее кол-во продуктов одинаково в обеих группах')
```

**Вывод**

На основе анализа данных с использованием теста Манна-Уитни при **уровне статистической значимости 5%** можно заключить, что **существуют статистически значимые различия** в количестве продуктов, которыми пользуются ушедшие и оставшиеся клиенты. Поскольку **p-value** оказалось **значительно меньше 0.05** (p-value = 4.33e-157), **мы отвергаем нулевую гипотезу** и подтверждаем, что ушедшие клиенты в среднем используют больше продуктов, чем оставшиеся.

Отношение средних значений **1.35** также показывает, что **ушедшие клиенты имеют на 34.9% больше продуктов** в использовании по сравнению с оставшимися, что подтверждает вывод о существовании значимой разницы между группами.

- **Активность клиентов**: Слабая корреляция (0.26) указывает на то, что активные клиенты также склонны к оттоку.
- **Пол клиентов**: Мужчины имеют выше отточность, в сравнении с женщинами 24% против 13%

----

2. **Выводы по выдвинутым гипотезам**

Анализ данных с помощью теста Манна-Уитни дал следующие результаты:

- **Количество продуктов**: Статистически значимые различия (p-value = 4.33e-157) подтверждают, что ушедшие клиенты используют больше продуктов.
- **Баланс**: Статистически значимые различия (p-value = 3.72e-42) показывают, что ушедшие клиенты имеют больше средний баланс.
- **Заработная плата**: Не выявлено статистически значимых различий (p-value = 1.0) в среднем уровне заработной платы между ушедшими и оставшимися клиентами.
- **Скоринговые баллы**: Обнаружены статистически значимые различия в средних скоринговых баллах между группами (p-value = 1.0), даже при отношении среднего значения скоринга ушедших и оставшихся в 1.02, что указывает на небольшое, но важное влияние скоринга.

Таким образом, ключевыми факторами оттока являются **количество продуктов и баланс**, тогда как **возраст, очки скоринга и наличие кредитной карты** имеют менее значительное влияние.

```python
df_segment_1 = df_filtered.loc[
    (
        ((df_filtered['age'] >= 20) & (df_filtered['age'] <= 35)) |
        ((df_filtered['age'] >= 51) & (df_filtered['age'] <= 60))
    ) &
    (df_filtered['balance'] > 900000) &
    (df_filtered['products'] >= 2)
]

# Уровень оттока в сегменте
churn_rate_segment_1 = df_segment_1['churn'].mean()

# Количество клиентов в сегменте
df_segment_1_count = df_segment_1['user_id'].count()

# Вывод информации об уровне оттока
print(f"Количество клиентов в сегменте: {df_segment_1_count}")
print(f"Уровень оттока в сегменте 1: {churn_rate_segment_1:.2f}")
print(f"Общий уровень оттока: {total_mean_churn:.2f}")
```

**Сегмент 3**: Клиенты с кредитной картой, высоким балансом и баллами собственности

 - Количество продуктов: Выше или равно 1

```python
# Фильтрация по признакам
df_segment_3 = df_filtered.loc[
    (df_filtered['products'] >= 1) &
    (df_filtered['balance'] >= 900000) &
    (df_filtered['equity'] >= 1) &
    (df_filtered['age'] >= 20) & (df_filtered['age'] <= 35)
]

# Уровень оттока в сегменте
churn_rate_segment_3= df_segment_3['churn'].mean()

# Количество клиентов в сегменте
df_segment_3_count = df_segment_3['user_id'].count()

# Вывод информации об уровне оттока
print(f"Количество клиентов в сегменте: {df_segment_3_count}")
print(f"Уровень оттока в сегменте 3: {churn_rate_segment_3:.2f}")
print(f"Общий уровень оттока: {total_mean_churn:.2f}")
```
